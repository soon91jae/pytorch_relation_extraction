```python
from matplotlib import pyplot as plt
import mplcursors
from pandas import DataFrame


df = DataFrame(
    [("Alice", 163, 54),
     ("Bob", 174, 67),
     ("Charlie", 177, 73),
     ("Diane", 168, 57)],
    columns=["name", "height", "weight"])

df.plot.scatter("height", "weight")
mplcursors.cursor().connect(
    "add", lambda sel: sel.annotation.set_text(df["name"][sel.target.index]))
plt.show()

# test: skip
```

```python
df = DataFrame(dict(x=[1.0,2.0,3.,4.], y=[1.,2.,3.,4.], label=['1','2','1','2'], sent = ['one','two','three','four']))
groups = df.groupby('label')
colormap = plt.cm.tab20c.colors 
fig, ax = plt.subplots(figsize=(9, 9))
i = 0
sent_dict = {'[3. 3.]': "three"}
for name, group in groups:
    #df.plot.scatter(group.x, group.y, marker='o', linestyle='', color = colormap[i], ms=12, label=name, markersize = 2)
    ax.plot(group.x, group.y, marker='o', linestyle='', color = colormap[i], ms=12, label=name, markersize = 2)
    i += 1
ax.legend()
mplcursors.cursor().connect(
    "add", lambda sel: sel.annotation.set_text(sent_dict[str(sel.target)]))
plt.show()
```
